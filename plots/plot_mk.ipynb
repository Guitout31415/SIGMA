```python
import shutup; shutup.please()
import scanpy as sc
import seaborn as sns
import anndata as ad
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (18, 8)
```

# At least 1 gene seen 1 time

## Load data and normalization

```python
# RAW DATA
adata_raw_zeng = sc.read_h5ad("/home/lemgui01/CellExtractor/data/raw_human/Zeng_biorxiv_2023_12_26_573390.h5ad")

adata_raw_park_fig1 = sc.read_h5ad("/home/lemgui01/CellExtractor/data/raw_human/Park_science_science_aay322_fig1.h5ad")
adata_raw_park_stem = sc.read_h5ad("/home/lemgui01/CellExtractor/data/raw_human/Park_science_science_aay322_stem.h5ad")
adata_raw_park = ad.concat([adata_raw_park_fig1, adata_raw_park_stem], label="study", keys=["Park_fig1", "Park_stem"], index_unique="-")

# CANDIDATE + TARGET COLUMN
adata_harmonized_zeng_11 = sc.read_h5ad("/home/lemgui01/CellExtractor/data/results_mk_1-1/harmonized/Zeng_biorxiv_2023_12_26_573390.h5ad")

adata_harmonized_park_fig1 = sc.read_h5ad("/home/lemgui01/CellExtractor/data/results_mk_1-1/harmonized/Park_science_science_aay322_fig1.h5ad")
adata_harmonized_park_stem = sc.read_h5ad("/home/lemgui01/CellExtractor/data/results_mk_1-1/harmonized/Park_science_science_aay322_stem.h5ad")
adata_harmonized_park_11 = ad.concat([adata_harmonized_park_fig1, adata_harmonized_park_stem], label="study", keys=["Park_fig1", "Park_stem"], index_unique="-")
```

```python
sc.pp.normalize_total(adata_raw_zeng, target_sum=1e6)
sc.pp.log1p(adata_raw_zeng)

sc.pp.normalize_total(adata_raw_park, target_sum=1e6)
sc.pp.log1p(adata_raw_park)

adata_raw_zeng_11 = adata_raw_zeng.copy()
adata_raw_zeng_12 = adata_raw_zeng.copy()
adata_raw_zeng_21 = adata_raw_zeng.copy()

adata_raw_park_11 = adata_raw_park.copy()
adata_raw_park_12 = adata_raw_park.copy()
adata_raw_park_21 = adata_raw_park.copy()
```

## Zeng

```python
# Marquage des cellules
adata_raw_zeng_11.obs["is"] = "Other"
adata_raw_zeng_11.obs.loc[adata_raw_zeng_11.obs_names.isin(adata_harmonized_zeng_11.obs_names), "is"] = "Candidate"
adata_raw_zeng_11.obs.loc[adata_raw_zeng_11.obs_names.isin(adata_harmonized_zeng_11.obs[adata_harmonized_zeng_11.obs["is_target"]].index), "is"] = "Target"

# Légende des types cellulaires
adata_raw_zeng_11.obs["cell_type_for_legend"] = "other"
mk_cell_type = adata_harmonized_zeng_11.obs["cluster"][adata_harmonized_zeng_11.obs["is_target"]].value_counts().loc[lambda x: x>0].index
for type in mk_cell_type:
    mask = adata_raw_zeng_11.obs["cell_type"].str.contains(type, case=False, na=False)
    adata_raw_zeng_11.obs.loc[mask, "cell_type_for_legend"] = adata_raw_zeng_11.obs.loc[mask, "cell_type"]

# UMAP
palette_pipeline = {"Other": "grey", "Candidate": "green", "Target": "red"}
palette_zeng = sns.color_palette("tab20")
palette_zeng[-9] = "grey"

fig, ax = plt.subplots(1, 2)
sc.pl.umap(adata_raw_zeng_11, color="is", ax=ax[0], show=False, size=30,
           palette=palette_pipeline, title="MKs found by pipeline")
sc.pl.umap(adata_raw_zeng_11, color="cell_type_for_legend", ax=ax[1], show=False, size=30,
           palette=palette_zeng, title="Clusters from Zeng et al.", legend_fontsize=8)
plt.tight_layout()
plt.subplots_adjust(wspace=0.25)
plt.savefig("/home/lemgui01/CellExtractor/plots/umap_mk_zeng_11.pdf", bbox_inches="tight", dpi=300)
plt.show()
```
