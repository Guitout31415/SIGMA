```python
import scanpy as sc
import matplotlib.pyplot as plt
```

```python
PATH_SIGMA = "/mnt/projects_tn03/AD_EB_singlecell/article/results_10xgenomics/merge.h5ad"

SEUIL = 0.95
```

## 1. Data loading

```python
adata_sigma = sc.read_h5ad(PATH_SIGMA)

print(f"adata_sigma : {adata_sigma.shape[0]:,} cells")
```

## 2. Utility functions

```python
def plot_A(ax):
    n, bins, _ = ax.hist(adata_sigma.obs["score"].dropna(), bins=10)
    ax.set_xlabel("Score")
    ax.set_ylabel("Number of cells")
    ax.set_yscale("log")
    ax.grid(axis='y', linestyle='--', alpha=0.3)
    ax.axvline(SEUIL, color='black', linestyle='--', linewidth=1)
    ax.text(SEUIL-0.085, ax.get_ylim()[1]*0.85, str(SEUIL),
            color='black', ha='left', va='top', fontsize=14)
    ax.set_title("Distribution of SIGMA scores", fontsize=16)
```

## Figure - Score distribution

```python
fig, ax = plt.subplots(figsize=(8, 5))
plot_A(ax)
fig.tight_layout()
fig.savefig("figures/10x-full.svg", bbox_inches="tight", dpi=300)
plt.show()
```
