```python
import scanpy as sc
import re
import matplotlib.pyplot as plt
```

```python
adata_sigma = sc.read_h5ad("/mnt/projects_tn03/Data_MK/article/results_10xgenomics/merge.h5ad")
```

# Score plot

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
ax.hist(adata_sigma.obs["score"].dropna(), bins=100)
ax.set_xlabel("Score")
ax.set_yscale("log")
ax.grid(axis='y', linestyle='--')
ax.set_ylabel("Number of cells")
ax.set_title("Distribution of SIGMA scores")

fig.savefig("figures/10x_Score.svg", bbox_inches="tight", dpi=500)
plt.show()
```
